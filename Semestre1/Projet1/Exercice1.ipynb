```python
import random #permet de générer des nombres aléatoires
import numpy as np #importation de la bibliothéque numpy pour effectuer des calculs numéiques
import matplotlib.pyplot as plt # importation du module pyplot de la bibliothéque matplotlib pour la visualisati
```

#### 1_Tirage de n variables de y

```python
def generer(a,b,n):
    yi = np.random.uniform(low = a, high = b, size =[1, n] )  
    return yi

print("Les réalistations {y1,...,yn} des variables{Y1,...,Y2} sont:", generer(0,10,10))
    
    
```

#### 2_Calcul de la moyenne empirique

```python
def moyenne_empirique(a,b, n):
    return np.mean(generer(a,b,n))

print("La moyenne est:", moyenne_empirique(0,10,10))
```

#### 3_ Répétition de l'expérience pour N= 5000

```python
def generer_moy(a,b,n,N):
    listemoy = []
    for i in range(N):
        yi = np.random.uniform(low = a, high = b, size = [1, n] )  
        listemoy.append(np.mean(yi))
    return listemoy
        

print("La moyenne des 5000 répétitons de l'expérience est:", generer_moy(0, 10,10, 5000))
```

```python
#4_Construction de l'histogramme

simulation = generer_moy(0, 10,10,5000) #appel de la fonction generer()
plt.rcParams["figure.figsize"] = [16,9] #paramétrage du fenêtre graphique utliser rcParams
plt.hist(simulation, density = True, label="simulation") # simulation: valeur d'entrée contenant les arguments de la loi Uniforme,edgecolor permet de distinguer chaque bande de l'histogramme, label = "simulation" place le mot dans la légende,densité: Si Truele premier élément du tuple de retour sera le nombre uniforme pour former une densité de probabilité 
plt.xlabel("valeurs des moyennes empiriques")
plt.ylabel("valeurs générées")
plt.title("Histogramme des 5000 répétitions") #affiche le titre
plt.draw() #efface les messages au dessus du graphique
#plt.savefig("Histogramme des 5000 répétitions.png", bbox_inches = "tight")# bbox = "tight" permet de bien proportionner le graphique lors de l'enregistrement
```
